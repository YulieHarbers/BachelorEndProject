```python
labels = np.array([])

rows = summary_df.shape[0]


for i in range(rows):
    firstvideo = summary_df.iloc[i]["first_file"]
    secondvideo = summary_df.iloc[i]["second_file"]
    thirdvideo = summary_df.iloc[i]["third_file"]
    partnr = summary_df.iloc[i]["participant"]
    # Convert video numbers to integers (automatically strips leading zeros)
    firstvideo = int(firstvideo)
    secondvideo = int(secondvideo)
    thirdvideo = int(thirdvideo)

    try:
        label1 = dataS.loc[partnr][f"affEng{firstvideo}"]
        label2 = dataS.loc[partnr][f"affEng{secondvideo}"]
        label3 = dataS.loc[partnr][f"affEng{thirdvideo}"]
        label1 = int(label1)
        label2 = int(label2)
        label3 = int(label3)
        mean_label = (label1+label2+label3)/3
        binary_label = 1 if mean_label > 0.5 else 0
        labels = np.append(labels,binary_label)
        print(label1, label2, label3, mean_label,binary_label)
    except KeyError as e:
        labels = np.append(labels, np.NaN)
        print(f"Can't find video number {firstvideo} , {secondvideo} OR {thirdvideo} for participant {e}")

summary_df["labels"] = labels
summary_df["labels"] = summary_df["labels"].astype('Int64')
summary_df.to_csv(r'C:\Users\20223868\OneDrive - TU Eindhoven\Documents\School\year 3\BEP\code\fau_with_labels_3.csv', index=False)
```
