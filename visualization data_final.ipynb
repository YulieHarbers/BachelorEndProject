```python

melted = df.melt(id_vars='labels', value_vars=hrv_features,
                     var_name='Feature', value_name='Value')
rename_dict = {
    'MeanHR_3_prev_diff_3': "MeanHR_3_prev_diff_3",
    'RMSSD_3_mean_diff_3': "RMSSD",
    'MeanNN_3_mean_diff_3': "MeanNN",
    'SDNN_3_mean_diff_3': "SDNN",
    'SDRMSSD_3_mean_diff_3': "SDRMSSD",
    'LF/HF_3_mean_diff_3': "LF/HF"

}

melted['Feature'] = melted['Feature'].replace(rename_dict)
# Create faceted boxplots (one per HRV feature)
g = sns.catplot(
    data=melted,
    x='labels', y='Value',
    col='Feature',
    kind='point',
    col_wrap=3,
    sharey=False,       # allow different y-axes
    height=4,           # height of each subplot
    aspect=0.9          # aspect ratio
)

g.set_titles("{col_name}", size=10, fontweight='bold')
g.set_axis_labels("label", "Value")
g.set_xticklabels(['Not Engaged (0)', 'Engaged (1)'])
for ax in g.axes.flatten():
    ax.tick_params(axis='x', labelrotation=15)
    ax.tick_params(axis='y', labelsize=8)

plt.subplots_adjust(top=0.9)
g.fig.suptitle("Point plots of HRV Features by Engagement", fontsize=14, fontweight="bold")
plt.show()
```
